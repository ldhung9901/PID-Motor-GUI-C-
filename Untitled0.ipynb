```
import numpy as np
import cv2
import matplotlib.pyplot as plt
# Read the image in greyscale
img = cv2.imread('puppy.jpg',0)

#Iterate over each pixel and change pixel value to binary using np.binary_repr() and store it in a list.
def tachmatphang(img):
    lst = []
    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            lst.append(np.binary_repr(img[i][j] ,width=8)) # width = no. of bits
    eight_bit_img = (np.array([int(i[0]) for i in lst],dtype = np.uint8) * 128).reshape(img.shape[0],img.shape[1])
    seven_bit_img = (np.array([int(i[1]) for i in lst],dtype = np.uint8) * 64).reshape(img.shape[0],img.shape[1])
    six_bit_img = (np.array([int(i[2]) for i in lst],dtype = np.uint8) * 32).reshape(img.shape[0],img.shape[1])
    five_bit_img = (np.array([int(i[3]) for i in lst],dtype = np.uint8) * 16).reshape(img.shape[0],img.shape[1])
    four_bit_img = (np.array([int(i[4]) for i in lst],dtype = np.uint8) * 8).reshape(img.shape[0],img.shape[1])
    three_bit_img = (np.array([int(i[5]) for i in lst],dtype = np.uint8) * 4).reshape(img.shape[0],img.shape[1])
    two_bit_img = (np.array([int(i[6]) for i in lst],dtype = np.uint8) * 2).reshape(img.shape[0],img.shape[1])
    one_bit_img = (np.array([int(i[7]) for i in lst],dtype = np.uint8) * 1).reshape(img.shape[0],img.shape[1])
    slice_plane=[]
    finalr = eight_bit_img+seven_bit_img+six_bit_img+five_bit_img+four_bit_img+three_bit_img+two_bit_img
    plane = (eight_bit_img,seven_bit_img,six_bit_img,five_bit_img,four_bit_img,three_bit_img,two_bit_img,one_bit_img)
    slice_plane=(one_bit_img,finalr)
    return slice_plane
slice_plane = tachmatphang(img)

fig = plt.figure()
test = np.array(img, dtype='uint8')
plt.imshow(test, cmap='gray')
plt.imshow(slice_plane[0], cmap='gray')
# Mã hóa thông tin và giấu tin:

#Mã hóa
import operator
def string2bin(string):
    return ''.join(f"{ord(i):08b}" for i in string)
    
def xormang(s,key):

  lens = len(s);chuoi="";
  for i in range(lens):
    chuoi=chuoi+str(operator.xor(int(s[i]),int(key[i])))
  
  return chuoi
def xuly(s,image):
    length=len(s);
    size=image.shape;row=size[0]
    k=-1;
    for i in range(row):
        a = size[1]-2; b=size[1]-1;
        k=k+1;
        if (k<length):
            image[i][a]=s[k];
            k=k+1;
            if(k<length):
                image[i][b]=s[k];
    return image

print("Nhập đoạn thông tin cần giấu: ")
string = input()
s=string2bin(string);

print("Nhập kí tự khóa: ")
key0= input()
key0= string2bin(key0)
key=""
for i in range(int(len(s)/8)):
  key = key+key0
s=xormang(s,key)
check='11111111';
s=s+check
size=slice_plane[0].shape;

if (len(s)<=2*size[0]):
    image = xuly(s,slice_plane[0])   
    new_image=image+slice_plane[1]
    fig = plt.figure()
    test = np.array(new_image, dtype='uint8')
    plt.imshow(test, cmap='gray')
else:
    print("Không thể mã hóa thông điệp bằng hình ảnh này !")


# Giải mã thông tin từ ảnh đã giấu:

img = new_image
slice_plane = tachmatphang(img)
imgxuly = slice_plane[0]
print("Nhập kí tự khóa: ")
password=input()
password=string2bin(password)
def xuly2(imgxuly,key0):
    sTest ='';
    k=0;s='';kq='';
    size=imgxuly.shape; row=size[0];
    a=size[1]-2; b=size[1]-1;
    for i in range(row-1):
        s=s+str(imgxuly[i][a]);
        k=k+1;
        s=s+str(imgxuly[i][b]);
        k=k+1;
        if (k==8):
            k=0;
            if (int(s,2)==255):
                break;
            # num = xormang(s,key0);
            # maascii=chr(int(num,2));
            # kq=kq+maascii;
            sTest += s ;
            s='';
    return sTest
if (password==key0):
  kq=xuly2(imgxuly,key0);
  key=""
  for i in range(int(len(kq)/8)):
    key = key+key0
  abc = xormang(kq,key);
  p="";
  f=0;
  alo='';
  for i in range(int(len(abc))):
      f+=1;
      p+=str(abc[i]);
      if (f==8):
         f=0;
         maascii=chr(int(p,2));
         
         alo=alo+maascii;
         p=''; 
  print(alo)  
  alo='';       
else:
    print("Nhập sai rồi !")
```
